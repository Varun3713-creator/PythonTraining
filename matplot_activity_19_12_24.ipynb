```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.figure()
plt.plot(x,y,color='green',linestyle='--',label='Simple line')
plt.title('Simple line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.xticks(x,['one','two','three','four','five'])
plt.yticks(x,['one','four','nine','sixteen','twnety five'])
plt.savefig('line_plot.png',dpi=300)
plt.show()
```

```python
#scatter plot
plt.figure()
plt.scatter(x,y,color='red',marker='o')
plt.title('Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
#bar plot
cat=['a','b','c','d']
val=[3,7,2,5]
plt.figure 
plt.bar(cat,val)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  

plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram sample')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("2x2 Grid:")
plt.subplot(2, 2, 1)  
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='green')
plt.title(" Plot 1")

plt.subplot(2, 2, 2)  
plt.plot(x, np.cos(x), color='red')
plt.title("Plot 2")

plt.subplot(2, 2, 3) 
plt.scatter(x, np.random.rand(100), color='blue')
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(1000), bins=30, color='purple', alpha=0.7)
plt.title("Plot 4")


plt.tight_layout()

plt.show()
```

```python
#Excercise
import matplotlib.pyplot as plt

cities = ['Mysuru', 'Mandya', 'Mangalore', 'Bangalore']
population = [200000, 80000, 300000, 800000]

plt.subplot(1, 2, 1)  
plt.plot(cities, population, color='blue', marker='*')
plt.title("Line Plot")
plt.ylabel("Population")
plt.xticks(rotation=45, fontsize=10)

plt.subplot(1, 2, 2)
plt.bar(cities, population, color='green', alpha=0.7)
plt.title("Bar Plot")
plt.ylabel("Population")
plt.xticks(rotation=45, fontsize=10)
plt.yticks(rotation=45, fontsize=10)
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')  


sns.lineplot(x="total_bill", y="tip", data=data)

plt.title("Line Plot Example")
plt.show()
```

```python
sns.barplot(x="day", y="tip", data=data)

plt.title("Bar Plot Example")
plt.show()
```
