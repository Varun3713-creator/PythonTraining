```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

# URL of the website to scrape
url = 'http://quotes.toscrape.com/'
```

```python
# Send a GET request to fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python

# Find all quotes on the page
quotes = soup.find_all('div', class_='quote')
quotes
```

```python
# Create lists to store extracted data
quote_texts = []
authors = []
tags = []
```

```python
# Loop through each quote and extract relevant data
for quote in quotes:
    # Get the quote text
    quote_text = quote.find('span', class_='text').text
    quote_texts.append(quote_text)
    
    # Get the author
    author = quote.find('small', class_='author').text
    authors.append(author)
    
    # Get the tags (if any)
    tag_list = quote.find('div', class_='tags').find_all('a', class_='tag')
    tags.append([tag.text for tag in tag_list])
```

```python
# Create a DataFrame to store the data
df = pd.DataFrame({
    'Quote': quote_texts,
    'Author': authors,
    'Tags': tags
})

# Display the DataFrame
print(df)

# Save the data to a CSV file
df.to_csv('quotes.csv', index=False)
```
